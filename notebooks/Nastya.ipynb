### Dataset description

The dataset belongs to the socio-economic domain and contains quality of life indicators for different countries between 2015 and 2024.

It includes information related to purchasing power, safety, healthcare, cost of living, pollution, climate conditions, and overall quality of life.

```python
import pandas as pd
```

```python
data = pd.read_csv("../data_sets/quality_of_life_indices_by_country.csv")
```

```python
data.head()
```

```python
data.shape
```

The dataset contains 12 fields.

There is one categorical field (Country), one ranking field (Rank), and ten variables describing different aspects of quality of life.

These variables include the overall Quality of Life Index, Purchasing Power Index, Safety Index, Health Care Index, Cost of Living Index, Property Price to Income Ratio, Traffic Commute Time Index, Pollution Index, Climate Index, and Year.

### Descriptive statistics

```python
statistics = pd.DataFrame({
    "Mean":[data["Quality of Life Index"].mean(),
            data["Purchasing Power Index"].mean(),
            data["Safety Index"].mean(),
            data["Health Care Index"].mean(),
            data["Cost of Living Index"].mean()],

    "Median":[
        data["Quality of Life Index"].median(),
        data["Purchasing Power Index"].median(),
        data["Safety Index"].median(),
        data["Health Care Index"].median(),
        data["Cost of Living Index"].median()],

    "Standard Deviation":
        [
        data["Quality of Life Index"].std(),
        data["Purchasing Power Index"].std(),
        data["Safety Index"].std(),
        data["Health Care Index"].std(),
        data["Cost of Living Index"].std()
        ],
},
    index=["Quality of Life Index",
           "Purchasing Power Index",
           "Safety Index",
           "Health Care Index",
           "Cost of Living Index"]
)

statistics.round(2)
```

Among the selected indicators, the Quality of Life Index has the highest variability, while the Health Care Index shows the lowest variability.

The mean and median values are relatively close for all indicators, suggesting that the data does not contain many extreme values.
